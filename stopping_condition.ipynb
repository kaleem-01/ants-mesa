```python
import numpy as np
np.unique(df_stop["state_counts"])
```

```python
import matplotlib.pyplot as plt
df_grouped = df_stop.groupby("prob_random")
```

```python
for name, group in df_grouped:
    print(group["prob_random"])
```

```python
for name, group in df_grouped:
    plt.hist(group["state_counts"], label=group["prob_random"].mean(), bins=2,)
    plt.legend()
    plt.show()
```
