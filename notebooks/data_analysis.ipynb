```python
def powerlaw_fit(X, Y):
    slope, y_intercept = np.polyfit(
        np.log(X),
        np.log(Y),
        deg=1
    )

    a = np.exp(y_intercept)
    b = slope

    return a, b
```

```python
results_pheromone_array = results_pheromone[results_pheromone["Step"] > 100]
```

```python
Y = np.array(results_pheromone["food_collected"]).flatten()
```

```python
N = len(Y)
Y_unique, Y_freq = np.unique(Y, return_counts=True)
Y_freq = Y_freq / N
Y_err = 1.96 * np.sqrt(Y_freq * (1 - Y_freq) / N)

X_fit = Y_unique[:9]
Y_fit = Y_freq[:9]

a, b = powerlaw_fit(X_fit, Y_fit)

Y_pred = a * Y_unique ** b

fig, ax = plt.subplots()

ax.errorbar(
    Y_unique,
    Y_freq,
    yerr=Y_err,
    label='Empirical Data'
)
ax.plot(Y_unique, Y_pred, label='Powerlaw Fit', color='black', linestyle='--')
ax.loglog()
ax.set_xlabel("Meal sizes")
ax.set_ylabel("Probability density")
ax.legend()
plt.show()
```

```python
df = pd.read_csv("../data/prob_random.csv")
df_grouped = df.groupby("prob_random")

mean = {}
std = {}

for name, group in df_grouped:
    mean[name] = group["Home üè†"].mean()
    std[name] = group["Home üè†"].std()

plt.plot(list(mean.keys()), list(mean.values()), marker='o')
plt.fill_between(list(mean.keys()), [mean[k] - std[k] for k in mean.keys()], [mean[k] + std[k] for k in mean.keys()], alpha=0.3)
plt.xlabel("Random Probability")
plt.ylabel("Home üè†")
plt.show()
```
